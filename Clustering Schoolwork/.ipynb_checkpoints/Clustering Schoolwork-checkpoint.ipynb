```python
%matplotlib inline
import numpy as np
import networkx as nx
from sklearn import datasets
from matplotlib import pyplot as plt
#加载鸢尾花数据集
iris = datasets.load_iris()
```

```python
iris_data = iris["data"]
iris_target = iris["target"]

def scatter_plot(dim1, dim2):  #dim1/dim2：iris_data中第几个数
    for t,marker,color in zip(range(3),"^os","rgb"):
        # zip()接受任意多个序列参数，返回一个元组tuple列表
        # 用不同的标记和颜色画出每种品种iris花朵的前两维数据
        #plt.xlim(xmax=8,xmin=0)
        #plt.ylim(ymax=8,ymin=0)
        plt.scatter(iris_data[iris_target==t, dim1], iris_data[iris_target==t, dim2], marker=marker,c=color)
        #iris_data[iris_target==t, dim1], iris_target==t遍历target中为t的所有data中dim1列      

plt.subplot(231)
scatter_plot(0,1)
plt.subplot(232)
scatter_plot(0,2)
plt.subplot(233)
scatter_plot(0,3)
plt.subplot(234)
scatter_plot(1,2)
plt.subplot(235)
scatter_plot(1,3)
plt.subplot(236)
scatter_plot(2,3)

plt.show()
```

```python
#用鸢尾花的第三、四维进行聚类，相似度用欧式距离表示，权值即欧式距离
#求距离( 相似度矩阵)
def juli(data1, data2):  #data:[x, y]
    x = data1[2]-data2[2]
    y = data1[3]-data2[3]
    dis = (x*x + y*y)**0.5
    dis = round(dis, 2)
    return dis
W = []
for i in iris_data:
    J_L = []
    for j in iris_data:
        jl = juli(i, j)
        J_L.append(jl)
    W.append(J_L)
W = np.array(W)
W = W.reshape(150,150)
W
```

```python
#散点图连线
k = 0.4
def scatter_plot(dim1, dim2):
    for t,marker,color in zip(range(3),"^os","rgb"):
        #plt.xlim(xmax=9,xmin=4)
        #plt.ylim(ymax=8,ymin=0)
        plt.scatter(iris_data[iris_target==t, dim1], iris_data[iris_target==t, dim2], marker=marker,c=color)
for i in range(150):
    for j in range(150):
        if W[i][j]<=k:
            plt.plot([iris_data[i][2],iris_data[j][2]],[iris_data[i][3],iris_data[j][3]])
        else:
            continue
scatter_plot(2,3) #散点图
plt.show()
```

```python
#求带权邻接矩阵A
A = np.zeros(150*150)
A = A.reshape(150,150)
for i in range(150):
    for j in range(150):
        if W[i][j]<=0.4:
            A[i][j] = W[i][j]
        else:
            A[i][j] = 0
A
```

```python
#求度数矩阵D
D = np.zeros(150*150)
D = D.reshape(150,150)
for a in range(150):
    for b in range(150):
        if a != b:
            D[a][b] = 0
        else:
            num = 0
            for m in range(150):
                num = num+A[a][m]
            num = round(num, 2)
            D[a][b] = num
D
```

```python
#求拉普拉斯矩阵L
L = np.zeros(150*150)
L = L.reshape(150,150)
L = D-A
L
```

```python
#求L的特征值w和特征向量v
w, v = np.linalg.eig(L)
#特征值排序
inds = np.argsort(w)[:3]
#将特征向量组成矩阵U
U = v[:, inds]
data = U
```

```python
#K-means
#计算距离
def juli(a, b):
    return np.sum(((a-b)**2)**0.5)
#随机质心
def cent(data):
    random_ind = np.random.choice(data.shape[0], size = 3)
    centers = [data[i] for i in random_ind]
    centers = np.array(centers)
    return centers
#多轮kmeans
def kmeans(data, k=3):
    x = False
    #检查质心与计算的中心是否相等
    def converged(center1, center2):
        set1 = set([tuple(c) for c in center1])
        set2 = set([tuple(c) for c in center2])
        x = (set1 == set2)
        return (set1 == set2)
    #初始化参数
    n = 150
    centers = cent(data)
    label = np.zeros(150, dtype=np.int)
    assement = np.zeros(150)
    converged = False
    
    while not converged:
        old_centers = np.copy(centers)
        for i in range(n):
            min_dist, min_index = np.inf, -1
            for j in range(k):
                dist = juli(data[i], centers[j])
                if dist<min_dist:
                    min_dist, min_index = dist, j
                    label[i] = j
            assement[i] = juli(data[i], centers[label[i]])**2
        
        for m in range(k):
            centers[m] = np.mean(data[label==m], axis=0)
        converged = x
    return centers, label, np.sum(assement)
```

```python
best_assement = np.inf
best_centroids = None
best_label = None
for i in range(10):
    centers, label, assement = kmeans(data,3)
    if assement < best_assement:
        best_assement = assement
        best_centroids = centroids
        best_label = label
data0 = data[best_label==0]
data1 = data[best_label==1]
data2 = data[best_label==2]
```
